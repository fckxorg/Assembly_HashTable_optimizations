```python
files = ["constant.log", "strlen.log", "ascii_len.log", "ascii_xor.log", "jenkins.log"]
data = []
for filename in files:
    cur_data = []
    with open(filename, "r") as file:
        for line in file:
            cur_data.append(int(line));
    data.append(cur_data)
```

```python
from matplotlib import pyplot as plt
```

```python
names = ["Constant", "Strlen", "ASCII/LEN", "ASCII_XOR_ROTATE", "Jenkins"]
indices = [x for x in range(997)]
plt.figure(figsize=(20, 10))
for cur_data, cur_name in zip(data, names):
    plt.bar(indices, cur_data, label=cur_name, width=1.0)

plt.legend(prop={'size': 25})
plt.xlabel("Lists", fontsize=25)
plt.ylabel("N_WORDS", fontsize=25)
plt.savefig("with_constant.png", format="png")
```

```python
indices = [x for x in range(3)]
data = [0.508, 1.684, 1.412, 1.337]
names = ["Optimized", "O0", "O1", "O3"]
bottom_names = [names[i] + " " + str(data[i]) + "s" for i in range(4)]

plt.figure(figsize=(20, 10))
for cur_bar, cur_name, cur_bname in zip(data, names, bottom_names):
    plt.bar(cur_bname, cur_bar, label=cur_name + " x" + str(round(cur_bar / data[0], 3)))

plt.legend(prop={'size': 25})
plt.ylabel("Time, seconds", fontsize=25)
plt.savefig("opt_comparison.png", format="png")
```
