```python
files = ["constant.log", "strlen.log", "ascii_len.log", "ascii_xor.log", "jenkins.log"]
data = []
for filename in files:
    cur_data = []
    with open(filename, "r") as file:
        for line in file:
            cur_data.append(int(line));
    data.append(cur_data)
```

```python
from matplotlib import pyplot as plt
```

```python
names = ["Constant", "Strlen", "ASCII/LEN", "ASCII_XOR_ROTATE", "Jenkins"]
indices = [x for x in range(997)]
plt.figure(figsize=(20, 10))
for cur_data, cur_name in zip(data[3:], names[3:]):
    plt.bar(indices, cur_data, label=cur_name)

plt.legend(prop={'size': 25})
plt.xlabel("Lists", fontsize=25)
plt.ylabel("N_WORDS", fontsize=25)
plt.savefig("without_len.png", format="png")
```
