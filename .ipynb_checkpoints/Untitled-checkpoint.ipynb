```python
hashes_stat = []
files = ["logs/constant.log", "logs/strlen.log", "logs/ascii_len.log", "logs/ascii_xor.log", "logs/jenkins.log"]
for el in files:
    data = []
    with open(el, "r") as file:
        for line in file:
            data.append(int(line))
    hashes_stat.append(data)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(20,10))    
for i in range(5):
    plt.plot(hashes_stat[i])

plt.show()
```
